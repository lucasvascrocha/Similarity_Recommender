# Pré processamento de dados faltantes

```python
# exploratory DF with % NA and types values
exploracao = pd.DataFrame({'nomes' : df.columns, 'tipos' : df.dtypes,
                           'NA #': df.isna().sum(), 'NA %' : (df.isna().sum() / df.shape[0]) * 100})
exploracao
```

```python
# % de dados faltantes, um corte abaixo de 5 % seria o ideal,visto que a maioria das features tem uma quantidade muito grande de dados faltantes
sns.distplot(exploracao['NA %'], bins=50)
```

```python
# dropando as features com Na > 5% e as linhas <= 5% Na
df = Clean_na().drop_col(5,df)
```


# # Início Engenharia de feature

```python
id_df = df['id'].copy()
df_reduced = df.drop('id', axis=1)
```

```python
df_reduced.head()
```

```python
#  exploração de frequência de todos os dados menos numéricos
for coluna in df_reduced.drop('idade_empresa_anos',axis=1):
    print(coluna)
    sns.countplot(x=coluna, data =df_reduced)
    plt.show()
```

Devido a diversos dados desbalanceados optei por fazer o balanceamento manual destas features

```python
# faz as alterações nas features para balanceá-las
df_colum_changed = Feature_engineering().change_columns(df_reduced)
```

```python
df_colum_changed.head()
```

```python
#  exploração de frequência de todos os dados menos numéricos
for coluna in df_colum_changed.drop('idade_empresa_anos',axis=1):
    print(coluna)
    sns.countplot(x=coluna, data =df_colum_changed)
    plt.show()
```
